```python
filename = '/content/drive/My Drive/DATA/age.csv'
import csv

f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
result1 = []
result2 = []
result3 = []
result4 = []
result5 = []

for row in data:
    if '서울특별시 종로구 혜화동(1111065000)' in row[0]:
        for i in row[3:]:
            result1.append(int(i.replace(',', '')))

    elif '서울특별시 동대문구 이문제1동(1123074000)' in row[0]:
        for i in row[3:]:
            result2.append(int(i.replace(',', '')))

    elif '서울특별시 동대문구 회기동(1123071000)' in row[0]:
        for i in row[3:]:
            result3.append(int(i.replace(',', '')))

    elif '서울특별시 중구 장충동(1114058000)' in row[0]:
        for i in row[3:]:
            result4.append(int(i.replace(',', '')))

    elif '서울특별시 중구 성북구(1129057500)' in row[0]:
        for i in row[3:]:
            result5.append(int(i.replace(',', '')))

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')
plt.plot(result1, 'blue', label='서울특별시 종로구 혜화동(1111065000)')
plt.plot(result2, 'orange', label='서울특별시 동대문구 이문제1동(1123074000)')
plt.plot(result3, 'green', label='서울특별시 동대문구 회기동(1123071000)')
plt.plot(result4, 'red', label='서울특별시 중구 장충동(1114058000)')
plt.plot(result5, 'purple', label='서울특별시 중구 성북구(1129057500)')
plt.title('행정구역')
plt.legend()
plt.show()
```
